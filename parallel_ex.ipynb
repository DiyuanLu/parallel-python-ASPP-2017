```python
import h5py
import numpy as np
import matplotlib.pyplot as plt

# load the data from weather.hdf5 here into a
# numpy array temps (temperatures)
filename = 'weather.hdf5'
f = h5py.File(filename, 'r')

print(f)
```

```python
for dset in (f.keys()):
    print(dset)
```

```python
data = f["temperature"][:8]
```

```python
data.shape
```

```python
np.arange(2010, 2017)
```

```python
plt.show()
```

```python
def make_plot(i):
    """
    Use plt.imshow() with the option interpolation='lanczos'
    to plot temps[i].



    """
    plt.figure()
    plt.imshow(data[i,...], interpolation="lanczos")
    
years = np.arange(2010, 2018)
for i, year in enumerate(years):
    make_plot(i)
    plt.title("Year %r" %year)
    plt.colorbar()
    plt.savefig('{}.png'.format(years[i]))
    
```
